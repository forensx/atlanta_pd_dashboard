```python
import pandas as pd
import numpy as np
import datetime
import json
```

```python
df = pd.read_csv('CSV/all_csv.csv')
```

```python
df
```

```python
dropped = df[['Lat', 'Long', 'UC2_Literal', 'Report Date']]
```

```python
dropped
```

```python
data = []
for row in dropped.itertuples(index = False):
    data.append({
        'COORDINATES': [row[0], row[1]],
        'TYPE': row[2],
        'DATE': datetime.datetime.strptime(row[3], '%m/%d/%Y').strftime("%Y-%m-%d %H:%M:%S")
    })
```

```python
with open('2019_data.json', 'w') as f:
    json.dump(data, f)
```
